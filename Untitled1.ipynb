```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
np.random.seed(42)
merchant_prices = {'Starbucks': np.nan, 'Shell Gas': np.nan, 'Woolworths': np.nan, 'Netflix': 199.00, 'Uber': np.nan, 'Amazon': np.nan, 'Spotify': 99.00, 'Checkers': np.nan, 'Gym Membership': 450.00, 'Restaurant XYZ': np.nan}
category_map = {'Starbucks': 'Coffee', 'Shell Gas': 'Gas', 'Woolworths': 'Groceries', 'Netflix': 'Subscription', 'Uber': 'Transport', 'Amazon': 'Shopping', 'Spotify': 'Subscription', 'Checkers': 'Groceries', 'Gym Membership': 'Subscription', 'Restaurant XYZ': 'Dining'}
merchants = list(merchant_prices.keys())
n = 150
start_date = datetime(2025, 1, 1)
merchant_col = np.random.choice(merchants, n)
random_amounts = np.round(np.random.uniform(5, 200, n), 2)
fixed_amounts = pd.Series(merchant_col).map(merchant_prices).values
final_amounts = np.where(np.isnan(fixed_amounts), random_amounts, fixed_amounts)
date_col = [start_date + timedelta(days=int(x)) for x in np.random.randint(0, 180, n)]
df = pd.DataFrame({'date': date_col, 'merchant': merchant_col, 'amount': final_amounts})
df = df.sort_values('date').reset_index(drop=True)
df['category'] = df['merchant'].map(category_map)
df.to_csv('transactions.csv', index=False)
print(df.head())
summary = df.groupby('category')['amount'].sum().sort_values(ascending=False)
print(summary)
summary.plot(kind='bar', color='teal')
plt.title('Spending by Category')
plt.tight_layout()
plt.show()
recurring = df.groupby('merchant').agg(count=('amount', 'count'), avg_amount=('amount', 'mean'), std_amount=('amount', 'std')).reset_index()
recurring['likely_subscription'] = (recurring['count'] >= 3) & (recurring['std_amount'].fillna(0) < recurring['avg_amount'] * 0.15)
subscriptions = recurring[recurring['likely_subscription']].sort_values('avg_amount', ascending=False)
print(subscriptions[['merchant', 'count', 'avg_amount']])
```
